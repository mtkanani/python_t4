```python
#piechat
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10))
y=np.array([10,20,30,40,50])
plt.pie(y,labels=["a","b","c","d","e"],startangle=90,explode=[0,0.2,0,0.2,0],shadow=True,autopct="%1.1f%%")
plt.legend(title="alphabets")

plt.show()
```

```python
#create a  pie chart for popularity data of diffrent languages and siplat as pie chart for python
#pyhton 29
#java 19
#js 8
#php 6
#r 3
#create an exploded view of python show pr in each programming languesga of pie chart
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,10))

y=np.array([29,19,8,7,6,5,3])
plt.pie(y,labels=["pyt","java","js","c","php","c++","r"],shadow=True,autopct="%1.1f%%",explode=[0,0.2,0,0.2,0,0,0])
plt.legend(title="popularity chart")
plt.tight_layout()
plt.title("chat")
plt.show()
```
