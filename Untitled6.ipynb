```python
import matplotlib.pyplot as plt
a=[1,2,3,3,3,3,4,2,5,5,7,7,5,6,8,8,8,9]
plt.hist(a,align='right')
plt.title('histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt
a=[1,2,3,3,3,3,4,2,5,5,7,7,5,6,8,8,8,9]
plt.hist(a,bins=4,color='red')
plt.title('histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
mylabels=['a','b','c','d']
plt.pie(x,labels=mylabels,explode=[0.1,0,0,0],autopct='%0.1f%%')
plt.legend()
plt.show()
# ,colors=['pink','blue','red','green']
```

```python
import matplotlib.pyplot as plt
catagories=['engineering','manufacturing','sales','profit']
costs=[1.35,3.60,2.25,1.80]
plt.pie(costs,labels=catagories,explode=[0,0,0,0.1],autopct='%0.1f%%')
plt.legend(loc='upper left')
plt.show()
```
